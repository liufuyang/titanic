```python
import pandas as pd
import pylab as plt
%matplotlib inline
import numpy as np

df=pd.read_csv('data/train.csv', sep=',')
```

```python
import sklearn.linear_model as Linear_Model
LogisticRegression = Linear_Model.LogisticRegression()
```

```python
df['Sex_'] = df['Sex'].apply(lambda x: 0 if x=='female' else 1)
msk = np.random.rand(len(df)) < 0.8
_train = df[msk]
_validation = df[~msk]
```

```python
LogisticRegression.fit(_train['Sex_'][:,np.newaxis],_train['Survived'])
```

```python
predict = LogisticRegression.predict(_validation['Sex_'][:,np.newaxis])
```

```python
correct_rate = 1-abs(_validation['Survived']-predict).sum()/float(len(_validation))
```

```python
print correct_rate
```

```python
test = pd.read_csv('data/test.csv', sep=',')
test['Sex_'] = test['Sex'].apply(lambda x: 0 if x=='female' else 1)
```

```python
test['Survived'] = LogisticRegression.predict(test['Sex_'][:,np.newaxis])
```

```python
import csv as csv
predictions_file = open("gendermodel.csv", "wb")
predictions_file_object = csv.writer(predictions_file)
predictions_file_object.writerow(["PassengerId", "Survived"])	# write the column headers
for index, row in test.iterrows():									# For each row in test file,
    predictions_file_object.writerow([row['PassengerId'], row['Survived']])			# write the PassengerId, and predict 1
predictions_file.close()
```
